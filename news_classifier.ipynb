```python
# Check karte hain sab libraries sahi import ho rahi hain
import torch
import transformers
import datasets
import sklearn
import pandas
import numpy

print("Torch:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
print("Transformers:", transformers.__version__)
print("Datasets:", datasets.__version__)
print("Scikit-learn:", sklearn.__version__)
print("Pandas:", pandas.__version__)
print("All imports successful!")
```

```python
# Hugging Face se AG News dataset load kar rahe hain
# Ye dataset 4 categories mein news headlines classify karta hai:
# 0 = World, 1 = Sports, 2 = Business, 3 = Sci/Tech
from datasets import load_dataset

dataset = load_dataset("ag_news")

# Dataset ka structure dekhte hain (train aur test split kitne rows ke hain)
print(dataset)
```

```python
# Naya path format use kar rahe hain (namespace/name)
from datasets import load_dataset

dataset = load_dataset("fancyzhx/ag_news")

print(dataset)
```

```python
# Dataset ka ek sample dekhte hain - pehli 3 rows
print(dataset["train"][0])
print(dataset["train"][1])
print(dataset["train"][2])
```

```python
# Poora dataset bohat bada hai (120,000 rows), isliye ek chota, balanced sample lete hain
# taake training fast ho aur laptop bhi handle kar sake

# Dataset ko shuffle karte hain (taake categories mix ho jayein, order mein na rahein)
shuffled_train = dataset["train"].shuffle(seed=42)
shuffled_test = dataset["test"].shuffle(seed=42)

# Chota sample select karte hain: 3000 training rows, 600 test rows
small_train = shuffled_train.select(range(3000))
small_test = shuffled_test.select(range(600))

print("Training samples:", len(small_train))
print("Testing samples:", len(small_test))

# Check karte hain sample mein categories ka distribution kaisa hai
import pandas as pd
train_df = pd.DataFrame(small_train)
print("\nCategory distribution in training sample:")
print(train_df['label'].value_counts().sort_index())
```

```python
# BERT ka tokenizer load kar rahe hain
# Ye tokenizer text ko numbers (tokens) mein convert karega jo BERT samajh sake
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

# Test karte hain ek sample text pe
sample_text = small_train[0]["text"]
print("Original text:", sample_text)

tokens = tokenizer(sample_text)
print("\nTokenized output:", tokens)
```

```python
# Pure dataset (train aur test dono) ko tokenize karte hain
# padding=True -> chote sentences ko bada karta hai (taake sab same length ke hon)
# truncation=True -> bohat lambe sentences ko chota kar deta hai (max 128 tokens tak)
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

# .map() function har row pe tokenize_function apply karta hai
tokenized_train = small_train.map(tokenize_function, batched=True)
tokenized_test = small_test.map(tokenize_function, batched=True)

print("Tokenization complete!")
print(tokenized_train[0].keys())
```

```python
# BERT ka classification model load kar rahe hain
# num_labels=4 -> kyunki humare paas 4 categories hain (World, Sports, Business, Sci/Tech)
from transformers import BertForSequenceClassification

model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=4)

# Model ko GPU pe bhejte hain (taake fast training ho)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

print("Model loaded successfully!")
print("Model is on device:", device)
```

```python
# Dataset ko PyTorch tensors ke format mein convert karte hain
# taake model isse directly samajh sake training ke dauran
tokenized_train.set_format("torch", columns=["input_ids", "attention_mask", "token_type_ids", "label"])
tokenized_test.set_format("torch", columns=["input_ids", "attention_mask", "token_type_ids", "label"])

print("Dataset format set to PyTorch tensors!")
print(tokenized_train[0])
```

```python
# Evaluation metrics ka function banate hain
# Ye training ke dauran accuracy aur F1-score calculate karega
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    # predictions mein probabilities hoti hain har category ke liye
    # argmax se hum sabse zyada probability wali category select karte hain
    predictions = np.argmax(predictions, axis=1)
    
    acc = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='weighted')
    
    return {"accuracy": acc, "f1": f1}

print("Evaluation function ready!")
```

```python
# Training ki settings define karte hain
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",           # model checkpoints yahan save hongi
    eval_strategy="epoch",            # har epoch ke baad evaluate karo
    save_strategy="epoch",            # har epoch ke baad model save karo
    learning_rate=2e-5,               # model kitni tezi se seekhta hai (chota rakhte hain BERT ke liye)
    per_device_train_batch_size=8,    # ek time pe kitne examples GPU process karega (6GB GPU ke liye 8 safe hai)
    per_device_eval_batch_size=8,
    num_train_epochs=3,               # pura dataset kitni baar dekhega model (3 baar)
    weight_decay=0.01,                # overfitting rokne ke liye
    logging_dir="./logs",
    logging_steps=50,                 # har 50 steps pe progress dikhao
    load_best_model_at_end=True,      # sabse acha model automatically select karo
)

print("Training arguments set!")
```

```python
# Trainer setup karte hain - ye actual training ko manage karega
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    compute_metrics=compute_metrics,
)

print("Trainer ready! Ab training start kar sakte hain.")
```

```python
# Training start kar rahe hain
# Ye process teen baar (3 epochs) pura training data dekhega aur seekhega
trainer.train()
```

```python
# Model ko test karte hain ek naye headline pe
def predict_category(text):
    labels = ["World", "Sports", "Business", "Sci/Tech"]
    
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)
    
    model.eval()
    with torch.no_grad():
        outputs = model(**inputs)
        prediction = torch.argmax(outputs.logits, dim=1).item()
    
    return labels[prediction]

# Test karte hain kuch examples pe
test_headlines = [
    "Apple unveils new iPhone with revolutionary AI features",
    "Manchester United wins dramatic match in final minute",
    "Stock markets crash amid economic uncertainty",
    "NASA discovers new exoplanet in habitable zone"
]

for headline in test_headlines:
    prediction = predict_category(headline)
    print(f"'{headline}' → {prediction}")
```

```python
%%writefile app.py
import streamlit as st
import torch
from transformers import BertTokenizer, BertForSequenceClassification

# Page ka title set karte hain
st.set_page_config(page_title="News Topic Classifier", page_icon="📰")

st.title("📰 News Topic Classifier (BERT)")
st.write("Koi bhi news headline likho, model batayega ye kis category ki hai!")

# Model aur tokenizer ko load karte hain (sirf ek baar, cache karke - taake baar baar load na ho)
@st.cache_resource
def load_model():
    model_path = "C:/bert_project_output"
    tokenizer = BertTokenizer.from_pretrained(model_path)
    model = BertForSequenceClassification.from_pretrained(model_path)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    return tokenizer, model, device

tokenizer, model, device = load_model()

# Labels define karte hain
labels = ["World", "Sports", "Business", "Sci/Tech"]

# User se input lete hain
headline = st.text_area("News headline yahan likho:", placeholder="e.g. Apple launches new AI-powered iPhone")

if st.button("Predict Category"):
    if headline.strip() == "":
        st.warning("Pehle koi headline likho!")
    else:
        inputs = tokenizer(headline, return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)
        with torch.no_grad():
            outputs = model(**inputs)
            probs = torch.softmax(outputs.logits, dim=1)
            prediction = torch.argmax(probs, dim=1).item()
            confidence = probs[0][prediction].item()

        st.success(f"**Predicted Category:** {labels[prediction]}")
        st.write(f"**Confidence:** {confidence*100:.2f}%")

        # Sab categories ki probability bhi dikhate hain
        st.write("### All Category Probabilities:")
        for i, label in enumerate(labels):
            st.write(f"{label}: {probs[0][i].item()*100:.2f}%")
```

```python
%%writefile app.py
import streamlit as st
import torch
from transformers import BertTokenizer, BertForSequenceClassification
import pandas as pd

st.set_page_config(page_title="News Topic Classifier", page_icon="📰", layout="centered")

# ---- Custom styling ----
st.markdown("""
    <style>
    .main-title {
        font-size: 2.2rem;
        font-weight: 700;
        margin-bottom: 0px;
    }
    .subtitle {
        color: gray;
        font-size: 1rem;
        margin-bottom: 25px;
    }
    .result-box {
        padding: 20px;
        border-radius: 10px;
        background-color: #f0f2f6;
        margin-top: 15px;
    }
    </style>
""", unsafe_allow_html=True)

st.markdown('<p class="main-title">📰 News Topic Classifier</p>', unsafe_allow_html=True)
st.markdown('<p class="subtitle">Fine-tuned BERT model that classifies news headlines into World, Sports, Business, or Sci/Tech.</p>', unsafe_allow_html=True)

# ---- Load model (cached so it only loads once) ----
@st.cache_resource
def load_model():
    model_path = "C:/bert_project_output"
    tokenizer = BertTokenizer.from_pretrained(model_path)
    model = BertForSequenceClassification.from_pretrained(model_path)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()
    return tokenizer, model, device

with st.spinner("Loading model..."):
    tokenizer, model, device = load_model()

labels = ["World", "Sports", "Business", "Sci/Tech"]
label_emojis = {"World": "🌍", "Sports": "⚽", "Business": "💼", "Sci/Tech": "🔬"}

# ---- Sidebar info ----
with st.sidebar:
    st.header("About")
    st.write("This app uses a BERT model fine-tuned on the AG News dataset to classify news headlines into 4 categories.")
    st.write("**Categories:**")
    for label in labels:
        st.write(f"{label_emojis[label]} {label}")
    st.write("---")
    st.write(f"**Running on:** {'GPU 🚀' if device.type == 'cuda' else 'CPU'}")

# ---- Main input ----
headline = st.text_area(
    "Enter a news headline:",
    placeholder="e.g. Apple unveils new AI-powered iPhone",
    height=100
)

col1, col2 = st.columns([1, 4])
with col1:
    predict_clicked = st.button("Predict", type="primary", use_container_width=True)

if predict_clicked:
    if headline.strip() == "":
        st.warning("Please enter a headline first.")
    else:
        with st.spinner("Analyzing..."):
            inputs = tokenizer(headline, return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)
            with torch.no_grad():
                outputs = model(**inputs)
                probs = torch.softmax(outputs.logits, dim=1)
                prediction = torch.argmax(probs, dim=1).item()
                confidence = probs[0][prediction].item()

        st.markdown('<div class="result-box">', unsafe_allow_html=True)
        st.markdown(f"### {label_emojis[labels[prediction]]} Predicted Category: **{labels[prediction]}**")
        st.write(f"Confidence: **{confidence*100:.1f}%**")
        st.markdown('</div>', unsafe_allow_html=True)

        st.write("")
        st.write("**Probability breakdown:**")
        prob_df = pd.DataFrame({
            "Category": labels,
            "Probability": [probs[0][i].item() for i in range(len(labels))]
        }).sort_values("Probability", ascending=False)

        st.bar_chart(prob_df.set_index("Category"))

# ---- Example headlines ----
st.write("---")
st.write("**Try an example:**")
examples = [
    "NASA discovers new exoplanet in habitable zone",
    "Manchester United wins dramatic match in final minute",
    "Stock markets surge after interest rate cut",
    "Floods displace thousands in southern region"
]
example_cols = st.columns(2)
for i, example in enumerate(examples):
    with example_cols[i % 2]:
        st.caption(f"• {example}")
```
